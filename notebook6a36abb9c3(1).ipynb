```python
def age(age):
    if age>60:
        return 1
    elif age>40 and age<=60:
        return 2
    elif age>0 and age<=40:
        return 3
    else:
        return 4
def chol(chol):
    if chol>300:
        return 1
    elif chol>200 and chol<=300:
        return 2
    elif chol>100 and chol<=200:
        return 3
    else:
        return 4
```

```python
df['age']=df['age'].apply(age)
df['chol']=df['chol'].apply(chol)
```
